```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,8))
plt.hist(df['Age'])
plt.title('AGE',fontsize=16,color='green')
plt.xlabel('X values',fontsize=14)
plt.ylabel('Y values',fontsize=14)
plt.xticks
plt.yticks
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df['BMI'],df['Glucose'],s=30,c='red',marker='*')
plt.title('BMI vs Glucose',fontsize=16,color='blue')
plt.xlabel('BMI',fontsize=14,color='green')
plt.ylabel('Glucose',fontsize=14,color='green')
plt.xticks
plt.yticks
```
